Create an interactive map showing climate data using Folium.

```python

import folium
import geopandas as gpd
from folium.plugins import MarkerCluster
from branca.colormap import linear  # Ensure this import is included

# TODO: Convert the projected GeoDataFrame to WGS84 for use with Folium
tz_wgs84 = gdf.to_crs(epsg=4326)

# TODO: Create an interactive map using Folium
def create_interactive_map(gdf, column, popup_columns, title, center=None, zoom_start=6):
    """Create an interactive map using Folium.
    
    Parameters:
    gdf (GeoDataFrame): The GeoDataFrame to map (must be in WGS84)
    column (str): The column to use for coloring
    popup_columns (list): Columns to include in the popup
    title (str): The title of the map
    center (tuple): The center coordinates [lat, lon]
    zoom_start (int): The initial zoom level
    
    Returns:
    folium.Map: The created interactive map
    """
    # TODO: Implement this function
    # Set map center to Tanzania if not provided
    if center is None:
        center = [-6.369028, 34.888822]  # Approximate center of Tanzania
    
    # Initialize the map
    folium_map = folium.Map(location=center, zoom_start=zoom_start)
    
    # Create a color scale
    colormap = linear.YlOrRd_09.scale(gdf[column].min(), gdf[column].max())
    
    # Add a Choropleth Layer
    folium.Choropleth(
        geo_data=gdf,
        data=gdf,
        columns=['REGION_NAM', column],
        key_on='feature.properties.REGION_NAM',
        fill_color='YlOrRd',
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name=f'{title} ({column})',
        highlight=True,
        reset=True
    ).add_to(folium_map)

    # Add Markers with Popups
    marker_cluster = MarkerCluster().add_to(folium_map)

    for _, row in gdf.iterrows():
        popup_text = "<br>".join([f"{col}: {row[col]}" for col in popup_columns])
        folium.Marker(
            location=[row.geometry.centroid.y, row.geometry.centroid.x],
            popup=folium.Popup(popup_text, max_width=300),
            tooltip=row['REGION_NAM']
        ).add_to(marker_cluster)
    
    return folium_map

# Create the interactive map
interactive_map = create_interactive_map(
    gdf=tz_wgs84, 
    column="ANNUAL_AVG_TEMP_C", 
    popup_columns=["REGION_NAM", "ANNUAL_AVG_TEMP_C", "ANNUAL_PRECIP_MM"], 
    title="Tanzania Climate Data"
)

# Display the interactive map
interactive_map
```

### Task 4.1: Analyze Temperature Trends
Analyze the trends in temperature across Tanzania over time.

```python
from scipy.stats import linregress
# TODO: Calculate temperature trends for each region
climate_df = tz_climate

def calculate_temperature_trends(df, region_column, year_column, temp_column):
    """Calculate temperature trends for each region.
    
    Parameters:
    df (DataFrame): The DataFrame containing climate data
    region_column (str): The column containing region identifiers
    year_column (str): The column containing year information
    temp_column (str): The column containing temperature values
    
    Returns:
    DataFrame: A DataFrame containing trend information for each region
    """
    # TODO: Implement this function using linear regression or other trend analysis methods
    pass
    trend_data = []

    # Group by region
    for region, group in df.groupby(region_column):
        group = group.sort_values(year_column)  # Ensure data is sorted by year

        # Perform linear regression (year vs. temperature)
        slope, intercept, r_value, p_value, std_err = linregress(group[year_column], group[temp_column])

        trend_data.append({
            region_column: region,
            "Temp_Trend": slope,  # Temperature trend (change per year)
            "P_Value": p_value,  # Statistical significance
            "R_Squared": r_value**2  # Goodness of fit
        })

    return pd.DataFrame(trend_data)

# Example usage
temp_trends = calculate_temperature_trends(climate_df, "REGION_COD", "YEAR", "ANNUAL_AVG_TEMP_C")

# Display results
print(temp_trends)
# TODO: Visualize the temperature trends
def plot_temperature_trends(trends_df, region_column, trend_column, title, figsize=(12, 8)):
    """Plot temperature trends by region.
    
    Parameters:
    trends_df (DataFrame): The DataFrame containing trend information
    region_column (str): The column containing region identifiers
    trend_column (str): The column containing trend values
    title (str): The title of the plot
    figsize (tuple): The figure size
    
    Returns:
    matplotlib.figure.Figure: The created figure
    """
    # TODO: Implement this function
    # Sort data by trend values for better visualization
    trends_df = trends_df.sort_values(by=trend_column, ascending=False)

    # Create the bar plot
    plt.figure(figsize=figsize)
    sns.barplot(x=trend_column, y=region_column, data=trends_df, palette="coolwarm")

    # Add labels and title
    plt.xlabel("Temperature Trend (°C per year)")
    plt.ylabel("Region")
    plt.title(title)

    plt.axvline(0, color="black", linestyle="--", linewidth=1)  # Reference line at zero trend

    plt.show()

# Example usage
plot_temperature_trends(temp_trends, "REGION_COD", "Temp_Trend", "Temperature Trends by Region")


```
